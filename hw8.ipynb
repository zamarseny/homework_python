```python
#Рассчитать t-test статистики для 3-4 переменных
```

```python
stats.ttest_ind(wines['Flavanoids'], wines['Proanthocyanins'])
```

```python
#средние показатели проантоцианидина и флавонидов не равны
```

```python
stats.ttest_rel(wines['Flavanoids'], wines['Proanthocyanins']) 
#stats.ttest_rel(wines['Flavanoids']/2.029270, wines['Proanthocyanins']/1.590899) 
```

```python
#зависимость переменных отвергается. 
#Если же каждую переменную разделить на среднее, то т-тест покажет зависимость с большой вероятностью
```

```python
stats.ttest_1samp(wines['Alcohol'] - wines['Proline'], 0)
#stats.ttest_1samp(wines['Alcohol']/13.000618 - wines['Proline']/746.893258, 0)
```

```python
#содержания спирта и пролина в вине неодинаковы
```

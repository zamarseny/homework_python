```python
#Т.е. у нас 774 вакансии на 39 страницах. Запускам цикл, чтобы скачать все страницы с 774 вакансиями Газпрома

vac = []
for i in tqdm_notebook(range(0, 1)):
    vac.append(requests.get("https://api.hh.ru/vacancies?employer_id=91", params={'page': i, 'per_page':20}).json())
```

```python
#Я сначала скачиваю ID всех вакансий Газпрома

pac=[]
#for i in range(0, 1):
for j in range(0, 19):
        pac.append(vac[0]['items'][j]['alternate_url'])
```

```python
import regex as re
```

```python
lili = [re.sub(r'[^0-9]', '', e) for e in pac]
```

```python
vah = []
#for i in tqdm_notebook(range(0, 39)):
for j in tqdm_notebook(range(0, 19)):
    vah.append(requests.get("https://api.hh.ru/vacancies?employer_id=91", params={'page': 0, 'per_page':20}).json()['items'][j]['alternate_url'])
```

```python
lulu = [re.sub(r'[^0-9]', '', e) for e in vah]
vak_url = 'https://api.hh.ru/vacancies/{}'
```

```python
var = []
for i in lulu:
    var.append(requests.get(vak_url.format(i)).json())
```

```python
df = pd.DataFrame(var)
```

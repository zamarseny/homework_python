## 7. Сделать кросс-валидацию данных с использованием подхода K-fold (n_folds=3)

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, y_train, y_test = train_test_split(data, Y, \
                                                        test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.ensemble import RandomForestClassifier
#from sklearn.datasets import make_regression

forestcl = RandomForestClassifier(max_depth=15, n_estimators=100, max_features='auto')
forestcl.fit(data_train, y_train)
```

```python
scores = cross_val_score(forestcl, data_train, y_train, cv=3)#, scoring='accuracy'
print(scores)
print(scores.mean())
```

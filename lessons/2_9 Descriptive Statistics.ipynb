```python
import random
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import kurtosis
from scipy.stats import skew

import seaborn
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
def coin():
    heads = 0
    for i in range(100):
        if random.random() <= 0.5:
            heads +=1
    return heads

def simulate(n):
   trials = []
   for i in range(n):
       trials.append(coin())
   return(sum(trials)/n)
```

```python
simulate(10)
```

```python
simulate(100)
```

```python
simulate(1000)
```

```python
simulate(1000000)
```

```python
mu=0
sigma=1
x=np.arange(-5,5,0.1)

y=stats.norm.pdf(x,mu,sigma)

plt.plot(x,y)
plt.title('Нормальное: mu=%i sigma=%.2f' % (mu,sigma), fontsize=15)
plt.xlabel('x')
plt.ylabel ('Вероятность')
```

```python
n = 10
p = 0.3
k = np.arange(0,21)

binom = stats.binom.pmf(k,n,p)
binom
```

```python
plt.plot(k,binom)
plt.title('Биномиальное: n=%i p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Количество успешных')
plt.ylabel ('Вероятность успеха')
```

```python
binom_sim = stats.binom.rvs(n=10, p=0.3, size=10000)
print (np.mean(binom_sim))
print (np.std(binom_sim,ddof=1))

plt.hist(binom_sim,bins=10,normed=True)
```

```python
rate=2
n=np.arange(0,10)
y=stats.poisson.pmf(n,rate)
y
```

```python
plt.plot(n,y)
plt.title('Пуассона: h=%i' % rate, fontsize=15)
plt.xlabel('Количество')
plt.ylabel ('Вероятность')
```

```python
data = stats.poisson.rvs(mu=2, loc=0, size=10000)
print (np.mean(data))
print (np.std(data,ddof=1))

plt.hist(data,bins=10,normed=True)
```

```python
a=0.5
b=0.5
x=np.arange(0.01, 1, 0.01)

y=stats.beta.pdf(x,a,b)

plt.plot(x,y)
plt.title('Бета: a=%i b=%.2f' % (a,b), fontsize=15)
plt.xlabel('x')
plt.ylabel ('Вероятность')
```

```python
x=np.arange(0,15,0.1)
 
y=stats.expon.pdf(x)

plt.plot(x,y)
plt.title('Экспоненциальное:',fontsize=15)
plt.xlabel('x')
plt.ylabel ('Вероятность')
```

```python
data=stats.expon.rvs(scale=2,size=10000)
print (np.mean(data))
print (np.std(data,ddof=1))

plt.hist(data,bins=10,normed=True)
```

```python
def stdev(nums):
    diffs = 0
    avg = sum(nums)/len(nums)
    for n in nums:
        diffs += (n - avg)**(2)
    return (diffs/(len(nums)-1))**(0.5)
```

```python
print(stdev(proline))
```

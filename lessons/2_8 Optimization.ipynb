```python
optimize.minimize(f, [2, -1], method="CG")
```

```python
def jacobian(x):
    return np.array((-2*.5*(1 - x[0]) - 4*x[0]*(x[1] - x[0]**2), 2*(x[1] - x[0]**2)))
```

```python
print(jacobian.shape())
```

```python
optimize.minimize(f, [2,-1], method="Newton-CG", jac=jacobian, options={'xtol': 1e-8, 'disp': True})
```
